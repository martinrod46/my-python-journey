```python
budget = 5000
monthly_spend = 1200

remaining = budget - monthly_spend
months_left = budget // monthly_spend

print(f"You have {remaining} left, which covers about {months_left} more months")

if months_left < 3:
    print("Time to be aggresive with job applications!")
else:
    print("I still have some breathing room!")
```

#### Loops

```python
job_sites = ["LinkedIn", "ai-jobs.com", "Indeed", "Wellfound"]

print(f"I am currently checking {len(job_sites)} job sites in order to find a job!")
print()

for site in job_sites:
    print(f"Checking {site} for new postings..")
    print(f"{site} has {len(site)} characters")
    print()
```

#### Functions


```python
def application_status(months_left):
    if months_left < 3:
        return "Time to be aggresive with job applications!"
    else: 
        return "I still have some breathing room!"

print(application_status(4))
print(application_status(2))
```
