#### Loops

```python
job_sites = ["LinkedIn", "ai-jobs.com", "Indeed", "Wellfound"]

print(f"I am currently checking {len(job_sites)} job sites in order to find a job!")
print()

for site in job_sites:
    print(f"Checking {site} for new postings..")
    print(f"{site} has {len(site)} characters")
    print()
```

#### Functions


```python
def application_status(months_left):
    if months_left < 3:
        return "Time to be aggresive with job applications!"
    else: 
        return "I still have some breathing room!"

print(application_status(4))
print(application_status(2))
```

```python
def application_status(name, months_left, min_months=3):
    if months_left < min_months:
        return f"{name}, time to be aggresive with job applications!"
    else:
        return f"{name}, you still have some breathing room."

print(application_status("Carlos", 2))
print(application_status("Carlos", 5))
print(application_status("Carlos", 2, min_months=5))
```

```python
candidates = [
    ("Carlos", 2),
    ("Maria", 5),
    ("Luis", 1)
]

for name, months_left in candidates:
    print(application_status(name, months_left))
```
